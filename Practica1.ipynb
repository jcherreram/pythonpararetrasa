```python
#Cuenta la cantidad de registros
import csv

with open('WorldCupMatches.csv') as archivoCSV: 
  lector = csv.reader(archivoCSV, delimiter=',')
  cantidadRegistros = 0
  for registro in lector:
    cantidadRegistros = cantidadRegistros + 1
    
print("Cantidad Registros: ", cantidadRegistros)
```

```python
#Muestra el primer registro (encabezados) y el largo del arreglo (que es lo mismo que decir: la cantidad de columnas que tiene el registro (encabezado)
import csv

with open('WorldCupMatches.csv') as archivoCSV: 
  lector = csv.reader(archivoCSV, delimiter=',')
  for registro in lector:
    print(registro)
    print(len(registro))
    break
```

```python
#Muestra el indice de cada columna
import csv

with open('WorldCupMatches.csv') as archivoCSV: 
  lector = csv.reader(archivoCSV, delimiter=',')
  for registro in lector:
    contador=0
    for columna in registro:
      print(contador,": ",columna)
      contador=contador+1
    break
```

```python
#Muestra "primer" registro (encabezado + primer registro)
import csv

with open('WorldCupMatches.csv') as archivoCSV: 
  lector = csv.reader(archivoCSV, delimiter=',')
  contador=0
  for registro in lector:
      contador=contador+1
      print(contador,": ",registro)
      if (contador>2):
        break
```
